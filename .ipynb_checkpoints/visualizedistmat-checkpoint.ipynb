```python
#map to taxids

import requests
import json

def unimap(terms, fromdb ,todb , verbose = False ):
    chunk = 100
    responses = ''
    mapping = {}
    for termchunk in [ terms[i*chunk:(i+1)*chunk] for i in range(int(len(terms)/chunk) +1) ]:
        url = 'http://www.uniprot.org/uploadlists/'
        params = {
        'from':fromdb,
        'to':todb,
        'format':'tab',
        'query':''.join([ t + ' ' for t in termchunk])
        }
        headers= {
            'User-Agent': 'Python dmoi@unil.ch''genus'
        }
        response = requests.get(url, params =params )
        if verbose == True:
            print(response.content.decode('utf-8'))
            
        lines = response.content.decode('utf-8').split('\n')
        Mapper = { line.split('\t')[0]:line.split('\t')[1] for line in lines[1:-1] }
        mapping.update(Mapper)
    return mapping

def uni2taxa(terms ):
    chunk = 100
    responses = ''
    mapping = {}
    for termchunk in [ terms[i*chunk:(i+1)*chunk] for i in range(int(len(terms)/chunk) +1) ]:
        url = 'http://www.uniprot.org/uploadlists/'
        params = {
        'from':'ACC+ID',
        'to':'ACC',
        'format':'tab',
        'columns': 'organism-id' ,
        'query':''.join([ t + ' ' for t in termchunk])
        }
        headers= {
            'User-Agent': 'Python dmoi@unil.ch''genus'
        }
        response = requests.get(url, params =params )
        lines = response.content.decode('utf-8').split('\n')
        Mapper = { line.split('\t')[1]:line.split('\t')[0] for line in lines[1:-1] }
        mapping.update(Mapper)
    return mapping

def uniparc2taxa(terms ):
    chunk = 100
    responses = ''
    mapping = {}
    for termchunk in [ terms[i*chunk:(i+1)*chunk] for i in range(int(len(terms)/chunk) +1) ]:
        url = 'http://www.uniprot.org/uploadlists/'
        params = {
        'from':'UPARC',
        'to':'ACC',
        'format':'tab',
        'columns': 'organism-id' ,
        'query':''.join([ t + ' ' for t in termchunk])
        }
        headers= {
            'User-Agent': 'Python dmoi@unil.ch''genus'
        }
        response = requests.get(url, params =params )
        lines = response.content.decode('utf-8').split('\n')
        Mapper = { line.split('\t')[1]:line.split('\t')[0] for line in lines[1:-1] }
        mapping.update(Mapper)
    return mapping

def ncbi2taxa(terms  ):
    chunk = 100
    responses = ''
    mapping = {}
    for termchunk in [ terms[i*chunk:(i+1)*chunk] for i in range(int(len(terms)/chunk) +1) ]:
        url = 'http://www.uniprot.org/uploadlists/'
        params = {
        'from':'P_REFSEQ_AC',
        'to':'ACC',
        'format':'tab',
        'columns': 'organism-id' ,
        'query':''.join([ t + ' ' for t in termchunk])
        }
        headers= {
            'User-Agent': 'Python dmoi@unil.ch''genus'
        }
        response = requests.get(url, params =params )
        lines = response.content.decode('utf-8').split('\n')
        Mapper = { line.split('\t')[1]:line.split('\t')[0] for line in lines[1:-1] }
        mapping.update(Mapper)
    return mapping


def ncbiGI2taxa(terms  ):
    chunk = 100
    responses = ''
    mapping = {}
    for termchunk in [ terms[i*chunk:(i+1)*chunk] for i in range(int(len(terms)/chunk) +1) ]:
        url = 'http://www.uniprot.org/uploadlists/'
        params = {
        'from':'P_ENTREZGENEID',
        'to':'ACC',
        'format':'tab',
        'columns': 'organism-id' ,
        'query':''.join([ t + ' ' for t in termchunk])
        }
        headers= {
            'User-Agent': 'Python dmoi@unil.ch''genus'
        }
        response = requests.get(url, params =params )
        lines = response.content.decode('utf-8').split('\n')
        Mapper = { line.split('\t')[1]:line.split('\t')[0] for line in lines[1:-1] }
        mapping.update(Mapper)
    return mapping



```
